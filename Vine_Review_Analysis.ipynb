```python
import pandas as pd
```

```python
vine_df=pd.read_csv("vine_table.csv", index_col=0)
vine_df
```

```python
#Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or 
#greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.

total_votes_df = vine_df.loc[vine_df["total_votes"] >= 20]
total_votes_df
```

```python
#Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the 
#rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

votes_perc_df =  total_votes_df.loc[total_votes_df["helpful_votes"] / total_votes_df["total_votes"] >= 0.5]
votes_perc_df
```

```python
#Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows 
#where a review was written as part of the Vine program (paid), vine == 'Y'.

vine_y_df = votes_perc_df.loc[votes_perc_df["vine"]=="Y"]
vine_y_df
```

```python
#Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.
vine_n_df = votes_perc_df.loc[votes_perc_df["vine"]=="N"]
vine_n_df
```

```python
#Determine the total number of reviews, the number of 5-star reviews, 
#and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

#total reviews
total_reviews = votes_perc_df["vine"].count()
total_reviews
```

```python
#five star with vine == Y
y_five_stars = vine_y_df.loc[(vine_y_df["star_rating"]== 5)]
y_five_stars.count()
```

```python
#five star with vine == N
n_five_stars = vine_n_df.loc[(vine_n_df["star_rating"]== 5)]
n_five_stars
```

```python
#total five star reviews
five_stars = votes_perc_df.loc[(votes_perc_df["star_rating"]== 5)]
five_stars
```

```python
#percentage of paid five star reviews
paid_reviews = (len(y_five_stars["vine"]) / len(five_stars["vine"]))*100
print(f'Percentage of Vine reviews with 5 stars is {paid_reviews}')

#percentage of paid five star reviews
unpaid_reviews = (len(n_five_stars["vine"]) / len(five_stars["vine"]))*100
print(f'Percentage Non-Vine reviews with 5 stars is {unpaid_reviews}')
```
